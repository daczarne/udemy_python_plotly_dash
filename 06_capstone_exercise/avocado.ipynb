```python
import pandas as pd
import plotly.express as px
```

```python
# Read the avocado data set
avocado = pd.read_csv("avocado.csv")
avocado
```

```python
# Filter the data set
filtered_df = avocado[avocado["geography"] == "Albany"]
```

```python
# Plot evolution of avg price
fig = px.line(
    filtered_df,
    x="date",
    y="average_price",
    color = "type",
    title='Avg price of avocados'
)
fig.show()
```
