```python
import pandas as pd
import plotly.express as px
```

```python
avocado = pd.read_csv("avocado.csv")
avocado
```

```python
filtered_df = avocado[avocado["geography"] == "Albany"][avocado["type"] == "conventional"]
filtered_df
```

```python
fig = px.line(filtered_df, x="date", y="average_price", title='Avg price of avocados')
fig.show()
```
