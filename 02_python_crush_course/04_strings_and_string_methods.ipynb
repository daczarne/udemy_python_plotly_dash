# Strings and string methods

String in Python can be initialized with either single quotes `'` or double quotes `"`

```python
print("Hello world")
print('Hello Python')
```

We can use the `+` operator to concatenate strings. We can use the `*` operator to repeat a string a number of times.

```python
question = "What are you studying?"
answer = "Plotly Dash"
print(question + " " + answer)
print(question + "\n" + answer)
print(question * 3)
```

The `len` function will give us the length of a string in Python. Every character will be counted, including spaces and punctuation.

```python
print(len(question))
print(len(answer))
```

In Python, string are 0-indexed arrays, so we can use the `[` operator to access them. We call it **indexing** when we access one character of the string. We call it **slicing** when accessing multiple ones. If we want them in reverse order we use negative numbers, starting from -1.

```python
print(answer)
print(answer[0])
print(answer[2])
print(answer[-1])
print(answer[len(answer) - 1])
```

When slicing, we use the following syntax: `var_name[start:stop:step]`. `step` is optional. If none is provided, Python will assume it to be 1. `start` and `stop` can be omited too. If we omit `start`, Python will assume that we want to start at index 0, if we omit `stop`, Python will assume we want to stop at the end of the string.

```python
print(answer) # the entire string
print(answer[2:6]) # from 2 to 5
print(answer[2:]) # from 2 to the end
print(answer[:4]) # form beginning to 3
print(answer[::2]) # all the string, but by 2
print(answer[:]) # The entire string
print(answer[::]) # The entire string
```

Methods are functions that are associated with objects and classes. They are members of the class. We access methods with the `.` operator. If you press tab, Python will show you a list of all avaiable methods for an instance of that class.

```python
print(answer)
print(answer.upper()) # All upper case
print(answer.lower()) # All lower case 
print(answer.swapcase()) # Swaps cases
print(answer.replace("s", "z")) # Replaces a substring
```

The `find` method will return the position of the first occurance of a given sub-string. If none is found, it will return -1.

```python
print(answer)
print(answer.find("t"))
print(answer.find("Dash"))
```

Python supports formated strings (also known as **f strings**)

```python
name = "Joan"
print(f"hello, {name}")
print(f"hello, {name.upper()}")
```
