# Lists and tuples

Lists and tuples are orderd sequences that can hold values of mixed data types. Lists are declared by way of `list()` or `[]`, while tuples are declared by way of `tuple()` or `()`. The main difference is that tuples are inmutable. So we use them we declaring global variables shouldn't change during the programs execution. This way we prevent bugs. 

```python
print(type([1,2]))
print(type((1,2)))
```

Lists support indexing and slicing.

```python
list_1 = [1, 2, 3]
list_2 = [1, "hello", "test", True, 2.6]
print("list 1: ", list_1)
print("list 1, postion 0: ",list_1[0])
print("list 1, positions 0 to 1 ", list_1[:2])
print("list 2: ", list_2)
print("list 2, skipping elements ", list_2[::2])
```

Lists support the `len` function

```python
print(len(list_1))
print(len(list_2))
```

Lists can be concatenated with the `+` operator

```python
print(list_1 + list_2)
```

We can append elements to a list using the `append` method. (Append means: adding to the end of the list). We can use the `pop` method to remove an element from a list. `pop` removes the last element by default because its default value is -1. But we can change it to `pop` elements in other positions.

```python
list_1.append(10)
print(list_1)
print(list_1.pop())
print(list_1)
print(list_1.pop(0))
print(list_1)
```

Lists can be sorted using the `sort` method. Keep in mind that characters will be sorted based on the English alphabet.

```python
list_3 = [1, 3, 7, 4, 8, 0, 100]
print(list_3)
list_3.sort()
print(list_3)
```

```python
list_4 = ["a", "c", "z", "ñ"]
print(list_4)
list_4.sort()
print(list_4)
```

We can call the `min` and `max` functions on lists.

```python
print(min(list_3))
print(max(list_3))
print(min(list_4))
print(max(list_4))
```

We can mutate a list using its indeces.

```python
print(list_1)
list_1[0] = 100
list_1[1] = "hello"
print(list_1)
```

# Sets

Sets are unordered collections of unique elements. If we try to assign the same element twice to a set, Python will just ignore it.

```python
set_1 = {1, 3, 2, 2}
print(set_1)
print(type(set_1))
```

We can cast a list into a set by using the `set` function

```python
print(list_4)
set_2 = set(list_4)
print(set(list_4))
print(set_2)
print(type(set_2))
```

We can add elements to a set with the `add` method.

```python
print(set_1)
set_1.add(9)
print(set_1)
```

If we try to add an element that is already in the set, Python will ignore it

```python
print(set_2)
set_2.add("a")
print(set_2)
```

We can use **set operations** to compare sets. The `difference` function tells us which elements are on one set, but not the other. `intersection` will give us the elements that are in both sets. Lastly, `union` will give us all unique elements in either set.

```python
set_1 = {1, 2, 3, 4, 5}
print(set_1)
set_2 = {1, 2, 6, 7, 10}
print(set_2)
print("difference: ", set_1 - set_2)
print("intersection: ", set_1 & set_2)
print("union: ", set_1 | set_2)
```

# Dictionaries

Dictionaries are unorderd mappings in key-value pairs. We can define them by passing `"key": value` pairs to `{}` or by passing `key = value` pairs to the `dict` function.

```python
print({"x": 1, "y": "hello"})
print(dict(x = 1, y = "hello"))
```

We can acces values by its key name

```python
dict_1 = {"k1": 1, "k2": "hello", "k3": True}
print(dict_1)
print("k1 is", dict_1["k1"])
```

Dictionaries can take any data type as there value, even other dictionaries.

```python
dict_2 = {"z": 123, "x": [1, 2, 4], "y": {"y1": 1, "y2": True}}
print(dict_2)
```

We can now access the list by way of

```python
print(dict_2["x"])
```

We can call elements of that same list with multiple `[`

```python
print("second element of list x is", dict_2["x"][1])
print("y1 element in dict y is", dict_2["y"]["y1"])
```

We can assign new key value pairs as follows

```python
dict_2["w"] = (1, 2)
print(dict_2)
```

# Booleans

Booleans can only take one of two values: `True` or `False`. All comparison operations evaluate to either one (or `None`).

```python
print("1 < 2 is", 1 < 2) # less than -> True
print("3 <= 2 is", 3 <= 2) # less than or equal to -> False
print('10 == "10" is', 10 == "10") # equal to -> Flase
print("10 != 10 is", 10 != 10) # not equal to -> Flase
print("11 > 10 is", 11 > 10) # greated than -> True
print("11 >= 20 is", 11 >= 20) # greated than or equal to -> Flase
print("None == None is", None == None) # True
print("1 == None is", 1 == None) # False
```

Python also supports comparison operators: `and`, `or`, and `not`

```python
print("h" == "h" and 2 == 2)
print(1 == 1 or 2 == 6)
print(1 == 1 and not 2 == 6)
```

We can use the `in` operator to check whether an element `x` is in a string, set, list, or a dictionary. 

```python
print("x" in ["x", "s"])
print("data" not in "data science")
```

```python
phone_dict = {"Justin": 123}
print(phone_dict)
print("Justin" in phone_dict)
```
