# Functions

Building functions allow us to avoid copy and pasting code, making our code easier to maintain. Function definition syntax is as follows:

```
def function_name(input_argument):
    """
    docstring
    """
    # Function body
```

````python
def greet(name):
    """
    Greet the person
    """
    print(f"hello, {name}")
````

````python
greet("john")
greet("mary")
````

````python
def adding(a, b):
    return a + b
````

````python
# Positional arguments
x = adding(4, 5)
print(x)
````

````python
# Keyword arguments
x = adding(a = 4, b = 5)
print(x)
````

Default paramenters

````python
def adding(a = 1, b = 1):
    return a + b
````

````python
y = adding()
print(y)
````

````python
def absolute_value(k):
    if k >= 0:
        return k
    else:
        return -k
````

````python
z = absolute_value(10)
print(z)
````

````python
z = absolute_value(-10)
print(z)
````

# Modules and Packages

A **module** is another Python script with statements, function definitions, variables, classes, etc that we can re-use. Combinations of modules are strucured in packages.

```
package
  |
  |_ module_1
  |
  |_ module_2
  |
  |_ module_3
```

Packages need to be imported in the current Python session to be used. We can do this with the `import` statement.

````python
import numpy as np
````

````python
np.arange(3)
````
