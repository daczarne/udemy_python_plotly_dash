# If statements

Use `if` statements to make conditinal branches to the programs flow.

```python
if 2 == 2:
    print("this is true")
```

```python
weather = "rainy"
if weather == "sunny":
    print("let's go to the park!")
elif weather == "cloudy":
    print("maybe we can go to the park?")
elif weather == "snowing":
    print("let's go snowboarding")
else:
    print("let's stay at home")
```

# Loops

Loops are constructs that allows to repeat the same task a number of times. `for` loops are used when we know the number of times a task needs to be executed. `while` loops are used for when we need a task to be executed as long as a condition is `True`.

In Python, loops can only be executed over *iterable* objects. Lists, sets, strings, and dicts are all examples of such objects. Another iterable object is the `range` object.

```python
l1 = [1, 2, 3, 4, 5, 6]
for i in l1:
    print("i = ", i)
```

We can nest `if` statements and `for` or `while` loops 

```python
for number in l1:
    if number % 2 == 0:
        print(number, " is an even number")
    else:
        print(number, " is an odd number")
```

```python
for number in l1:
    if number % 2:
        print(number, " is an odd number")
    else:
        print(number, " is an even number")
```

```python
total = 0
for number in l1:
    total += number
print(total)
```

```python
s1 = "plotly dash"
for s in s1:
    print(s)
```

```python
s1 = "plotly dash"
for s in s1:
    print(s, end="")
```

```python
l2 = [(1, 2), (3, 4), (5, 6)]
for v1, v2 in l2:
    print(v1 + v2)
```

```python
d1 = {"key1": 1, "key2": 2}
for key in d1:
    print(key)
```

```python
for value in d1.items():
    print(value)
```

```python
for key, value in d1.items():
    print(key, "-", value)
```

```python
counter = 0
while counter < 10:
    print(counter)
    counter += 1
```

```python
counter = 0
while counter < 10:
    print(counter)
    counter += 2
```

```python
counter = 0
while True:
    if counter < 10:
        print(counter)
        counter += 1
    else:
        break
```

```python
for i in range(10):
    print(i)
```

```python
for i in range(2, 10):
    print(i)
```

```python
for i in range(0, 10, 2):
    print(i)
```

```python
cnt = 0
l1 = ["a", "b", "c", "d"]
for letter in l1:
    print(letter)
    cnt += 1
print("cnt = ", cnt)
```

```python
l1 = ["a", "b", "c", "d"]
for cnt, letter in enumerate(l1):
    print(letter)
print("cnt = ", cnt + 1)
```
