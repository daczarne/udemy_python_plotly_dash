# Overview of data types

Data types in Python

| Data type | Name           | Type  |
|:---------:|:--------------:|:-----:|
| Numeric   | Integers       | int   |
| Numeric   | Floating point | float |
| Text      | Strings        | str   |
| Boolean   | Boolean        | bool  |
| Sequences | Lists          | list  |
| Sequences | Tuple          | tuple |
| Mapping   | Dictionaries   | dict  |
| Sets      | Sets           | set   |

The `type` function will give us the data type of an object in Python.

```python
print(type(1))
print(type(1.))
```

Python supports all basic arithmetic operations

```python
print(1 + 1) # Addition
print(2 - 1) # Substraction
print(3 * 4) # Multiplication
print(5 / 6) # Division
print(5 // 6) # Floor division
print(7 ** 8) # Power
print(10 % 2) # Modulo
```

Assign vairables with the syntax `name = value`

```python
x = 10
print(x)
```

Variables can be reassigned in Python, even if that changes their type.

```python
x = 11
print(x)
x = "hello"
print(x)
```
